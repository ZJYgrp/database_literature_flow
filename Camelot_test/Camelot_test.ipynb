# Table precise extraction from PDF through Camelot

```python
import camelot
```

```python
import tkinter
```

```python
import pandas
```

```python
from PyPDF2 import PdfFileReader
```

```python
input1 = PdfFileReader(open("./JBC_17430899.pdf", "rb"))
```

```python
input1.getNumPages()
```

```python
tables = camelot.read_pdf('./JBC_17430899.pdf',pages='11',flavor='stream')
## "Stream" can be used to parse tables that have whitespaces between cells to simulate a table structure.
```

```python
tables
## Now, we have a TableList object called tables, which is a list of Table objects. 
## We can get everything we need from this object.
```

```python
tables[0]
```

```python
tables[0].df
```

```python
df = tables[0].df
```

```python
df
```

# The table has been successfully extracted, but not in perfect precision. 
# A minor trimming to get rid of the mis-extracted first line is needed.

```python
df1 = df.drop([0])
```

```python
df1
```

```python
new_header = df1.iloc[0] #grab the first row for the header
df2 = df1[1:] #take the data less the header row
df2.columns = new_header #set the header row as the df header
```

```python
df2
```

# For cases that have worse parsing accuracy, an alternative is the visual debugging function
https://camelot-py.readthedocs.io/en/master/user/advanced.html#

```python
camelot.plot(tables[0], kind='contour').show()
## By plotting the contour of parsed tables, borders of the table can be specified.
```

```python
camelot.plot(tables[0], kind='textedge').show()
## Textedges can also be visualized by specifying kind='textedge'.
```

```python
tables = camelot.read_pdf('./JBC_17430899.pdf', 
                          pages='11',
                          flavor='stream', 
                          table_areas=['5,695,500,100']
                          )
tables[0].df
## table_areas accepts strings of the form: x1,y1,x2,y2

## (x1, y1) -> top-left and (x2, y2) -> bottom-right in PDF coordinate space. 

## In PDF coordinate space, the bottom-left corner of the page is the origin, with coordinates (0, 0).
```

```python
outputpath= './JBC_17430899.csv'
df2.to_csv(outputpath,sep=',',index=False,header=True)
## Download the extracted table in a csv file
```
